```
import matplotlib.pyplot as plt
import pandas as pd
```

***Line Graphs***<br>
Plot the equation y = x^2 for x = -100 to 100

```
x = pd.Series(range(-100,100))
y = x*x
plt.plot(x, y, label = "hello")
plt.legend()
```

Usin the same range for x. plot y = x^3 - 21

```
x = pd.Series(range(-100,100))
y = x**3-21
plt.plot(x, y, label = "hi")
plt.legend()
```

Plot each person's age and height in a scatter plot. Age should be the x axis and height should be the y axis.

```
data = {"Name":["Carl","Carla","Sam","Joshua","Ben","Andy","Mindy","Betty"], "Age":["42","32","12","21","2","5","94","96"], "Gender":["male","female","female","male","male","male","female","female"]}

df = pd.DataFrame(data)
# print(df)
df.insert(3,"Height",[72,60,65,78,32,46,58,64], True)
# print(df)
x = df.Age
y = df.Height
plt.scatter(x,y, label = "fat")
plt.legend()
```

Change the symbols for males to a blue asterisk(*). and females to a pink circle("o").

```
males = df[df["Gender"] == "male"]
females = df[df["Gender"] == "female"]

plt.scatter(males.Age,males.Height,color="blue",marker="*", label = 'Male')
plt.scatter(females.Age,females.Height,color="pink",marker="o", label = 'Female')
plt.legend()
```
